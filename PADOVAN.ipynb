**La sucesión de Padovan similar a la serie de Fibonacci con estructura recursiva similar. La fórmula recursiva es**

P(n) = P(n-2) + P(n-3)

P(n) = P(n-2) + P(n-3)

## Codigo del ejercicio 

```python
#Función para calcular el número padovan P (n)
def padovan(n):
    if n < 3:
        return 1
    else:
        #calcula en n-simo numero de padovan usando recursividad
        return padovan(n - 2) + padovan(n - 3)
#llama la funcion padovan y le envia por parametro la cantidad de n 
serie_padovan = [padovan(i) for i in range(12)]
#imprime la secuencia 
print(serie_padovan)
```

```python
#se plotean los n valores con los terminos de la serie de padovan 
import seaborn as sns
sns.set(style="whitegrid")

ax = sns.barplot(x=list(range(12)), y=serie_padovan)
ax.set_title("SUCESIÓN DE PADOVAN")
ax.set_ylabel("ELEMENTOS DE LA SECUENCIA")
ax.set_xlabel("n")

plt.plot(serie_padovan, linewidth=4, color='black')
plt.show()
```
