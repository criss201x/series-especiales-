Los números de Perrin son los números de la siguiente secuencia de enteros. 
3, 0, 2, 3, 2, 5, 5, 7, 10, 12, 17, 22, 29, 39… 

En términos matemáticos, la secuencia p (n) de números de Perrin está definida por la relación de recurrencia 

P (n) = P (n-2) + P (n-3) para n> 2, 

#### **con valores iniciales:** 
    P (0) = 3, P (1) = 0, P (2) = 2. 

Si una función int per (int n) que devuelve p(n). Por ejemplo, si n = 0, entonces per() debería devolver 3. Si n = 1, debería devolver 0 Si n = 2, debería devolver 2. Para n> 2, debería devolver p (n-2 ) + p (n-3) 

A continuación se muestra una implementación recursiva simple de la fórmula anterior.


# Codigo del ejercicio

```python
def perrin(n):
    if n == 0:
        return 3
    elif n == 1:
        return 0
    elif n == 2:
        return 2
    else:
        return perrin(n - 2) + perrin(n - 3)

perrin_serie = [perrin(i) for i in range(12)]
print(perrin_serie)
```

```python
#se plotean los n valores con los terminos de la serie de perrin 
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

ax = sns.barplot(x=list(range(12)), y=perrin_serie)
ax.set_title("SUCESIÓN DE PERRIN")
ax.set_ylabel("ELEMENTOS DE LA SECUENCIA")
ax.set_xlabel("n")

plt.plot(perrin_serie, linewidth=4, color='black')
plt.show()
```
